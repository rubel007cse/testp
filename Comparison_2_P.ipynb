```python
import statsmodels.api as sm
model1=sm.tsa.statespace.SARIMAX(prod_1['Count'],order=(1, 1, 1),seasonal_order=(1,1,1,30))
model2=sm.tsa.statespace.SARIMAX(prod_2['Count'],order=(1, 1, 1),seasonal_order=(1,1,1,30))
results_1=model1.fit()
results_2=model2.fit()
```

```python
prod_1['Forecast']=results_1.predict(start=0,end=250,dynamic=True)
prod_2['Forecast']=results_2.predict(start=0,end=250,dynamic=True)
prod_1[['Count','Forecast']].plot(figsize=(12,8))
```

# Notes
The **SARIMAX** model didn't work at all as data has no seasonality.
The 2nd product named 'HP USB 3.1 পেনড্রাইভ - 32GB' was fitted better than the 1st product named 'মেনজ হাফ স্লিভ কটন টি শার্ট'. 
Reason behind this:
<li> The first product was not enough frequent like the 2nd one. The purchase frequency of the 2nd product was more frequent than the 1st one. </li>

The holt winter's method worked best among these models.

In this kind of cases we generally make a frequency of weekly or monthly or yearly. This was just an experiment. A fixed frequency does a better forcasting always.
